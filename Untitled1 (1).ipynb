```python
for column in df.columns:
    if df[column].dtype=="object":
        df[column].fillna("unkonwn",inplace=True)
    else:
        df[column].fillna(0,inplace=True)
```

```python
df
```

```python
min_salaries=[]
max_salaries=[]
for salary in df["Salary Estimate"]:
    salary_range=salary.split(" ")[0].split("-")
    min_salary=salary_range[0].replace("$"," ").replace("k","")
    max_salary=salary_range[0].replace("$"," ").replace("k","")
    min_salaries.append(min_salary)
    max_salaries.append(max_salary)
df["min_salary (k) "]=min_salaries
df["max_salary (k) "]=max_salaries
```

```python
df
```

```python
Job_Titles=[]
departmentes=[]
for i in df["Job Title"]:
    title_parts=i.split(",")
    if len(title_parts)==2:
        Job_Titles.append(title_parts[0])
        departmentes.append(title_parts[1])
    else:
        Job_Titles.append(title_parts[0])
        departmentes.append("unkowen")
df['Job Title'] = Job_Titles
df['departmentes']=departmentes
        
        
     
```
